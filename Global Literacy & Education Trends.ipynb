```python
import pandas as pd

df_adult_lit = pd.read_csv('https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true',storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_adult_lit.head()
```

```python
df_adult_lit.rename(columns={'entity': 'country'}, inplace=True)
df_adult_lit.rename(columns={'code': 'country_code'}, inplace=True)
df_adult_lit.rename(columns={'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99': 'adult_lit_percent'}, inplace=True)
df_adult_lit.head()
```

```python
df_youth_lit = pd.read_csv('https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true',storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_youth_lit.head()
```

```python
df_youth_lit.rename(columns={'entity': 'country'}, inplace=True)
df_youth_lit.rename(columns={'code': 'country_code'}, inplace=True)
df_youth_lit.rename(columns={'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m': 'youth_male_lit_percent'}, inplace=True)
df_youth_lit.rename(columns={'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f': 'youth_female_lit_percent'}, inplace=True)
df_youth_lit.rename(columns={'owid_region': 'region'}, inplace=True)
df_youth_lit.tail()
```

```python
df_total_illit = pd.read_csv('https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true',storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_total_illit.head()
```

```python
df_total_illit.rename(columns={'entity': 'country'}, inplace=True)
df_total_illit.rename(columns={'code': 'country_code'}, inplace=True)
df_total_illit.tail()
```

```python
df_gdp_per_capita = pd.read_csv('https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true',storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_gdp_per_capita.head()
```

```python
df_gdp_per_capita.rename(columns={'entity': 'country'}, inplace=True)
df_gdp_per_capita.rename(columns={'code': 'country_code'}, inplace=True)
df_gdp_per_capita.rename(columns={'ny_gdp_pcap_pp_kd': 'gdp_per_capita'}, inplace=True)
df_gdp_per_capita.rename(columns={'owid_region': 'region'}, inplace=True)
df_gdp_per_capita.sample(5)
```

```python
df_avg_yrs_schooling = pd.read_csv('https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true',storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_avg_yrs_schooling.sample(4)
```

```python
df_avg_yrs_schooling.rename(columns={'entity': 'country'}, inplace=True)
df_avg_yrs_schooling.rename(columns={'code': 'country_code'}, inplace=True)
df_avg_yrs_schooling.rename(columns={'mf_youth_and_adults__15_64_years__average_years_of_education': 'total_pop_avg_education_year'}, inplace=True)
df_avg_yrs_schooling.rename(columns={'owid_region': 'region'}, inplace=True)
df_avg_yrs_schooling.sample(5)
```

```python
df_literacy['illiteracy_percent'] = 100 - df_literacy['adult_lit_percent']
```

```python
df_literacy['youth_lit_gender_gap'] = df_literacy['youth_male_lit_percent'] - df_literacy['youth_female_lit_percent']
```

```python
df_gdp_schooling['gdp_per_schooling_year'] = df_gdp_schooling['gdp_per_capita'] / df_gdp_schooling['total_pop_avg_education_year']
```

```python
df_gdp_schooling['education_index'] = df_gdp_schooling['literacy_rate'] / df_gdp_schooling['total_pop_avg_education_year']
```

```python
df_literacy['youth_literacy_avg'] = (df_literacy['youth_male_lit_percent'] + df_literacy['youth_female_lit_percent'])/2
```

```python
#EDA -  Univariate Analysis: Understanding the Distributions

import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Distribution of Adult Literacy (df_literacy)
sns.histplot(df_literacy['adult_lit_percent'], kde=True, ax=axes[0], color='teal')
axes[0].set_title('Distribution of Adult Literacy %')

# 2. Distribution of GDP per Capita (df_gdp_schooling)
sns.boxplot(x=df_gdp_schooling['gdp_per_capita'], ax=axes[1], color='coral')
axes[1].set_title('Spread of GDP per Capita')

# 3. Literacy Gender Gap (df_literacy)
sns.violinplot(y=df_literacy['youth_lit_gender_gap'], ax=axes[2], color='mediumpurple')
axes[2].set_title('Youth Literacy Gender Gap')

plt.tight_layout()
plt.show()
```

```python
# Negative skew - majority of observations are clustered at the high end (90%–100%). This confirms that high literacy is the "norm" globally.

# Box plot - he large orange box represents the Interquartile Range (IQR). It is squashed toward the left, meaning most countries have a relatively low GDP per Capita

# Violin plot - The "Bulge": The widest part of the purple shape is around 0. This is good news—it means in the majority of places, there is little to no difference between male and female youth literacy.
#The Upward Spike: There is a long "neck" stretching up toward 50. This represents regions where a massive gender gap exists (where male literacy is up to 50 percentage points higher than female literacy).
#The Downward Tip: The small section below zero indicates rare cases where female literacy actually exceeds male literacy.
```

```python
#EDA - Bivariate Analysis: Relationships & Trends

# 1. GDP vs Literacy Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_gdp_schooling, x='gdp_per_capita', y='literacy_rate', hue='region', alpha=0.7)
plt.xscale('log') # Use log scale if GDP variance is huge
plt.title('GDP per Capita vs. Literacy Rate (Log Scale)')
plt.show()

# 2. Years of Education vs Literacy
plt.figure(figsize=(10, 6))
sns.regplot(data=df_gdp_schooling, x='total_pop_avg_education_year', y='literacy_rate', scatter_kws={'alpha':0.5})
plt.title('Avg Years of Education vs. Literacy Rate')
plt.show()

# 3. Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = df_gdp_schooling[['gdp_per_capita', 'literacy_rate', 'total_pop_avg_education_year', 'gdp_per_schooling_year']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Filtering for World data or specific regions over time
world_trend = df_illiteracy[df_illiteracy['country'] == 'World']

plt.figure(figsize=(12, 6))
sns.lineplot(data=world_trend, x='year', y='literacy_rate', marker='o', label='Global Literacy')
sns.lineplot(data=world_trend, x='year', y='illiteracy_rate', marker='o', label='Global Illiteracy')
plt.title('Global Literacy vs Illiteracy Trends Over Time')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
import pymysql

conn = pymysql.connect(
    host = "localhost",
    user = "root",
    password = "sql1234",
    database = 'global_literacy'
)
cursor = conn.cursor()
```

```python
cursor.execute('''create table literacy_rates
(
country VARCHAR(100),
country_code VARCHAR(100),
year INT,
adult_lit_percent DECIMAL(20,2),
youth_male_lit_percent DECIMAL(20,2),
youth_female_lit_percent DECIMAL(20,2),
region VARCHAR(100),
illiteracy_percent DECIMAL(20,2),	
youth_lit_gender_gap DECIMAL(20,2),
youth_literacy_avg DECIMAL(20,2),
PRIMARY KEY (country, year)
);''')
```

```python
cursor.execute('''create table illiteracy_population
(
country VARCHAR(100),
country_code VARCHAR(100),
year INT,
illiteracy_rate DECIMAL(20,2),
literacy_rate DECIMAL(20,2),
PRIMARY KEY (country, year)
);''')
```

```python
cursor.execute('''create table gdp_schooling 
(
country VARCHAR(100),
country_code VARCHAR(100),
year INT,
gdp_per_capita DECIMAL(20,2),
region VARCHAR(100),
literacy_rate DECIMAL(20,2),
total_pop_avg_education_year DECIMAL(20,2),
population_historical DECIMAL(20,2),	
gdp_per_schooling_year DECIMAL(20,2),
education_index DECIMAL(20,2),
PRIMARY KEY (country, year)
);''')
```

```python
# Now try to write the dataframe to SQL with the correct engine
from sqlalchemy import create_engine

conn.close()
conn = pymysql.connect(
    host = "localhost",
    user = "root",
    password = "sql1234",
    database = 'global_literacy'
)
cursor = conn.cursor()

# Create SQLAlchemy engine using the PyMySQL connection
engine = create_engine('mysql+pymysql://root:sql1234@localhost/global_literacy')
```

```python
df_literacy.to_sql("literacy_rates",if_exists="replace",con =engine)
```

```python
df_gdp_schooling.to_sql("gdp_schooling",if_exists="replace",con =engine)
```

```python
# literacy_rates
#1. Get top 5 countries with highest adult literacy in 2020.

query = 'select country, adult_lit_percent from literacy_rates where year = 2020 order by adult_lit_percent limit 5'
df_top5_adult_per = pd.read_sql(query, conn)
print(df_top5_adult_per)
```

```python
#2. Find countries where female youth literacy < 80%.

query = 'select country, youth_female_lit_percent from literacy_rates where youth_female_lit_percent<80 order by youth_female_lit_percent desc'
df_female_lit_80 = pd.read_sql(query, conn)
print(df_female_lit_80)
```

```python
#3. Average adult literacy per continent (owid region).

query = 'select region, avg(adult_lit_percent) from literacy_rates group by region order by 2 desc'
df_avg_adult_lit_per = pd.read_sql(query, conn)
print(df_avg_adult_lit_per)
```

```python
#illiteracy_population
#4. Countries with illiteracy % > 20% in 2000.

query = 'select country, illiteracy_rate from illiteracy_population where illiteracy_rate > 20 and year = 2000 order by illiteracy_rate asc'
df_illiterate_2000 = pd.read_sql(query, conn)
print(df_illiterate_2000)
```

```python
#5. Trend of illiteracy % for India (2000–2020).

query = '''select country, year, illiteracy_rate from illiteracy_population where country = 'India' and year between 2000 and 2020 order by year'''
df_india_2000_2020 = pd.read_sql(query, conn)
print(df_india_2000_2020)
```

```python
# 6.Top 10 countries with largest illiterate population in the last year.

query = '''select country, illiteracy_rate  from illiteracy_population where year = (select max(year) from illiteracy_population) order by illiteracy_rate desc limit 10'''
df_large_illiterate_pop = pd.read_sql(query, conn)
print(df_large_illiterate_pop)
```

```python
#gdp_schooling
#7. Find countries with avg_years_schooling > 7 and gdp_per_capita < 5000.

query = '''SELECT country, year, total_pop_avg_education_year, gdp_per_capita FROM gdp_schooling WHERE total_pop_avg_education_year > 7 AND gdp_per_capita < 5000'''
df_gdp_per_capita = pd.read_sql(query, conn)
print(df_gdp_per_capita)
```

```python
#8. Rank countries by GDP per schooling for the year 2020.

query = '''select country, year, gdp_per_schooling_year from gdp_schooling where year = 2020 order by gdp_per_schooling_year'''
df_gdp_per_schooling_2020 = pd.read_sql(query, conn)
print(df_gdp_per_schooling_2020)
```

```python
# 9. Find global average schooling years per year.

query = '''select year, avg(total_pop_avg_education_year) from gdp_schooling group by year order by 2 desc'''
df_global_avg_schooling = pd.read_sql(query, conn)
print(df_global_avg_schooling)
```

```python
#Join Queries 
# 10. List top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling(less than 6).

query = '''SELECT country, year, gdp_per_capita FROM gdp_schooling WHERE year = 2020 and total_pop_avg_education_year < 6 order by gdp_per_capita desc'''
df_2020_top_countries = pd.read_sql(query, conn)
print(df_2020_top_countries)
```

```python
#11 Show countries where the illiterate population is high despite having more than 10 average years of schooling.

query = '''select i.country, i.year, i.illiteracy_rate
from illiteracy_population i inner join gdp_schooling g on i.country = g.country and i.year = g.year
where total_pop_avg_education_year > 10 order by illiteracy_rate desc'''
df_top_illiteracy_pop = pd.read_sql(query, conn)
print(df_top_illiteracy_pop)
```

```python
#12.  Compare literacy rates and GDP per capita growth for a selected country over the last 20 years. (country of your choice)

query = '''select l.country, l.year, l.adult_lit_percent, g.gdp_per_capita
from literacy_rates l inner join gdp_schooling g on l.country = g.country and l.year = g.year
where l.country = 'India' and l.year <= (select max(year) from literacy_rates)'''
df_top_illiteracy_pop = pd.read_sql(query, conn)
print(df_top_illiteracy_pop)
```

```python
# 13. Show the difference between youth literacy male and female rates for countries with GDP per capita above $30,000 in 2020.

query = '''select l.country, l.year, (l.youth_male_lit_percent - l.youth_female_lit_percent) youth_male_female_diff, g.gdp_per_capita
from literacy_rates l inner join gdp_schooling g on l.country = g.country and l.year = g.year
where g.gdp_per_capita > 30000 and l.year = 2020 order by g.gdp_per_capita desc'''
df_youth_male_female_diff = pd.read_sql(query, conn)
print(df_youth_male_female_diff)
```
